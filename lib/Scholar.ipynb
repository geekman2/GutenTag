```python
import pymongo
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from IPython.display import display

db_client = pymongo.MongoClient('mongodb://159.203.187.28:27017')

random_seed = np.random.seed(1791)
data = db_client.data.fiction
docs = data.aggregate([{"$match":{"fandom":"Chuck"}},{"$sample":{"size":1000}}],allowDiskUse=True)
doc_list = list(docs)
```

```python
all_keys = set()
for d in doc_list:
    all_keys.update(d['trigram_frequency'].keys())

print "NUMBER OF UNIQUE TRIGRAMS", len(all_keys)
```

```python
stuffs = []
for doc in doc_list:
    frequency = doc["trigram_frequency"]
    stuff = {trigram: round(frequency[trigram] * 10000, 3) if trigram in frequency else 0. for trigram in all_keys}
    stuff["_id"] = doc["_id"]
    stuff["_url"] = doc["url"]
    stuff["genres"] = doc["genres"]
    stuffs.append(stuff)

print "NON-ZERO FREQUENCY TRIGRAM COUNT:", len(stuffs)
df = pd.DataFrame(stuffs)
```

```python
display(df)
```

```python
cluster_data = df.drop(["_id","_url","genres"],axis=1)
pca = PCA(n_components=500)
pca = pca.fit(cluster_data)
print cluster_data.shape
reduced_data = pd.DataFrame(pca.transform(cluster_data))
print reduced_data.shape
```

```python
from sklearn.cluster import MiniBatchKMeans, AgglomerativeClustering
#clusterer = MiniBatchKMeans(n_clusters = 20, random_state=np.random).partial_fit(reduced_data)
clusterer = AgglomerativeClustering(n_clusters = 50).fit(cluster_data)
```

```python
labels = clusterer.labels_
db_client.close()

distribution = {}
for i,label in enumerate(labels):
    if label in distribution:
        distribution[label].append(df.ix[i]._url)
    else:
        distribution[label] = [df.ix[i]._url]

for i in distribution:
    print "Cluster Number {} has {} members".format(i,len(distribution[i]))

for j,cluster in enumerate(distribution):
    print "CLUSTER NUMBER {} CONTAINS".format(j)
    for k in distribution[j]:
        print k
```
